# Password Strength Checker

```python
import string

def check_password_strength(password):
    # Flags for requirements
    has_upper = False
    has_lower = False
    has_digit = False
    has_special = False
    
    # Define special characters
    special_characters = string.punctuation
    
    # Loop through each character
    for char in password:
        if char.isupper():
            has_upper = True
        elif char.islower():
            has_lower = True
        elif char.isdigit():
            has_digit = True
        elif char in special_characters:
            has_special = True
    
    # Evaluate strength
    strength = 0
    if has_upper: strength += 1
    if has_lower: strength += 1
    if has_digit: strength += 1
    if has_special: strength += 1
    
    # Feedback
    print("\nPassword Analysis:")
    if not has_upper:
        print("- Add at least one uppercase letter")
    if not has_lower:
        print("- Add at least one lowercase letter")
    if not has_digit:
        print("- Add at least one digit")
    if not has_special:
        print("- Add at least one special character")
    
    # Strength rating
    if strength == 4 and len(password) >= 8:
        print("✅ Strong Password")
    elif strength >= 2 and len(password) >= 6:
        print("⚠️ Medium Password")
    else:
        print("❌ Weak Password")

# Example usage
user_password = input("Enter your password: ")
check_password_strength(user_password)
```
